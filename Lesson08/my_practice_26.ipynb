```python

import numpy as np
import keras
from keras.layers import Dense

resnet50_model = keras.applications.resnet50.ResNet50()
```

```python
resnet50_model.summary()
```

```python
from keras.preprocessing import image
import numpy as np

new_image = image.load_img('Data/Prediction/test_image_3.jpg', target_size=(224,224))
new_image = image.img_to_array(new_image)
new_image = np.expand_dims(new_image, axis=0)

from keras.applications.resnet50 import preprocess_input
new_image = preprocess_input(new_image)

result = resnet50_model.predict(new_image)
result
```

```python
from keras.applications.resnet50 import decode_predictions
decode_predictions(result, top=5)
```

```python
label = decode_predictions(result)
```
