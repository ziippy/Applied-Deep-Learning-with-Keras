```python
import numpy as np
from keras.applications.vgg16 import VGG16
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
```

```python
classifier = VGG16()
```

```python
new_image = image.load_img('Data/Prediction/man_in_suit.jpg', target_size=(224,224))
```

```python
# 이미지를 배열로 변환
transformed_image = image.img_to_array(new_image)
transformed_image.shape
```

```python
# 이미지를 4차원 형태로 전환
transformed_image2 = np.expand_dims(transformed_image, axis=0)
transformed_image2.shape
```

```python
# 이미지를 전처리
transformed_image3 = preprocess_input(transformed_image2)
transformed_image3
```

```python
# 예측 변수를 생성
y_pred = classifier.predict(transformed_image3)
y_pred
```

```python
# 이미지 모양을 확인
y_pred.shape
```

```python
# 대상 이미지의 라벨로 확률이 가장 큰 5개를 선택
from keras.applications.vgg16 import decode_predictions
decode_predictions(y_pred, top=5)
```

```python
# 예측을 사람이 읽을 수 있는 형태로 변환
label = decode_predictions(y_pred)
```

```python
decoded_label = label[0][0]
decoded_label
```
