```python
import numpy as np
data = np.loadtxt("./data/pima-indians-diabetes.csv", delimiter=",")
```

```python
X = data[:,0:8]
y = data[:, 8]
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation

# 케라스 모델을 리턴하는 함수 생성
def build_model():
    # 케라스 모델 구성
    model = Sequential()
    model.add(Dense(16, activation='relu', input_dim=8))
    model.add(Dense(8, activation='relu'))
    model.add(Dense(1))
    # 모델 컴파일
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    # 모델 리턴
    return model
```

```python
# 케라스 래퍼와 사이킷런을 사용해, 모델을 위한 사이킷런 인터페이스를 만들어보자
from keras.wrappers.scikit_learn import KerasRegressor
YourModel = KerasRegressor(build_fn= build_model, epochs=300, batch_size=5, verbose=1)
```

```python
# 교차 검증에 사용할 반복자 선언
from sklearn.model_selection import KFold

kf = KFold(n_splits=5)
```

```python
# 교차 검증 수행을 위해 cross_val_score 함수 호출
# X, y 에 대해 교차 검증을 수행
from sklearn.model_selection import cross_val_score

results = cross_val_score(YourModel, X, y, cv=kf)
```

```python
# 결과 출력
# print accuracy for each fold
n_folds = 5
for f in range(n_folds):
    print("Test accuracy at fold ", f+1, " = ", results[f])
print("\n")

# print overal cross validation acuuracy plus the standard deviation of the accuracies
print("Final Cross Validation Test Accuracy:", results.mean())
print("Standard Deviation of Final Test Accuracy:", results.std())
```
