```python
# 신경망을 활용한 당뇨병 진단
# 총 768명의 환자 정보, 각 환자에 관해 8가지 측정치, 총 2개의 클래스 (1: 당녀병 환자, 0: 당뇨병이 아닌 환자)
# 즉, binary classification 문제
# 입력 차원은 8
```

```python
import numpy as np
data = np.loadtxt("./data/pima-indians-diabetes.csv", delimiter=",")
```

```python
X = data[:,0:8]
y = data[:, 8]
```

```python
print(f'X.shape: {X.shape}, Y.shape: {y.shape}')
```

```python
from sklearn.model_selection import train_test_split
test_size = 0.3
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=0)
```

```python
print(f'X_train.shape: {X_train.shape}, y_train.shape: {y_train.shape}')
print(f'X_test.shape: {X_test.shape}, y_test.shape: {y_test.shape}')
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
```

```python
model = Sequential()
model.add(Dense(units=8, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(units=1, activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(X_train, y_train, batch_size=5, epochs=300, validation_data=(X_test, y_test))
```

```python
import matplotlib.pyplot as plt
import matplotlib

plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.legend(loc='upper right')
```

```python
model2 = Sequential()
model2.add(Dense(units=16, activation='relu', input_dim=X_train.shape[1]))
model2.add(Dense(units=8, activation='relu', input_dim=X_train.shape[1]))
model2.add(Dense(units=1, activation='sigmoid'))
model2.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history2 = model2.fit(X_train, y_train, batch_size=5, epochs=350, validation_data=(X_test, y_test))
```

```python
plt.plot(history2.history['loss'], label='loss')
plt.plot(history2.history['val_loss'], label='val_loss')
plt.legend(loc='upper right')
```

```python
model3 = Sequential()
model3.add(Dense(units=16, activation='relu', input_dim=X_train.shape[1]))
model3.add(Dense(units=8, activation='relu', input_dim=X_train.shape[1]))
model3.add(Dense(units=4, activation='relu', input_dim=X_train.shape[1]))
model3.add(Dense(units=1, activation='sigmoid'))
model3.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history3 = model3.fit(X_train, y_train, batch_size=5, epochs=400, validation_data=(X_test, y_test))
```

```python
plt.plot(history3.history['loss'], label='loss')
plt.plot(history3.history['val_loss'], label='val_loss')
plt.legend(loc='upper right')
```
